# 목적
* ```imageio``` 로 GIF 만들기

````python
import tensorflow as tf
import imageio
import numpy as np
import matplotlib.pyplot as plt
import IPython

%matplotlib inline
````

## 데이터

````python
(X, _), (_, _) = tf.keras.datasets.mnist.load_data()
X.shape
````

## GIF 그리기
* ```X```에서 무작위 이미지 가져다가 GIF파일을 만들어보자.

````python
# 파일 경로
gif_file = './gif/digits.gif'

with imageio.get_writer(gif_file, mode='I') as writer:
    for i in range(1000):
        image = X[np.random.randint(X.shape[0])]
        writer.append_data(image)
````

* GIF 화면에 나타내기

````python
IPython.display.Image(gif_file)
````
