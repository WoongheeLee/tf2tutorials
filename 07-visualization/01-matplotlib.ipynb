````python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
````

## 데이터 준비

````python
(X_train, _), (_, _) = tf.keras.datasets.mnist.load_data()
````

````python
X_train.shape
````

## 그림 그리기

### 한장만 그리기

* 28 x 28 크기의 60,000개 데이터가 준비 됨
* 무작위 이미지 한 장만 화면에 띄우자.
* 이미지를 화면에 보일 때는 ```plt.imshow()``` 사용

````python
# 인덱스 랜덤 샘플링
idx = np.random.randint(X_train.shape[0])

# 이미지를 그릴 도화지를 정의
plt.figure(figsize=(5,5), dpi=80)

# 이미지 그리기
plt.imshow(X_train[idx])

# 도화지 화면에 출력
plt.show()
````

### $(M\times N)$ 개 그리기

* ```plt.subplots()```를 쓰면 도화지에 여러 이미지를 나타낼 수 있다.
* ```fig```와 ```ax```를 리턴하는데, ```ax```를 ```plt```로 생각할 수 있다. 물론 메소드가 좀 다름

````python
# 이미지 개수 정하기
nrows, ncols = 10, 5

# 도화지 정의
fig, ax = plt.subplots(nrows, ncols, figsize=(5, 5), dpi=100)

# subplot은 루프를 돌며 그리면 편하다.
for i in range(nrows):
    for j in range(ncols):
        # 랜덤 이미지 뽑아서 그림 그리기
        ax[i][j].imshow(X_train[np.random.randint(X_train.shape[0])])
        # 축 이름 필요 없음
        ax[i][j].axis('off')
        
# 도화지에 그린 그림 화면에 나타내기
plt.show()
````
